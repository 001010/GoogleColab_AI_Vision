```python
import numpy as np
import time

# YOUR TURN: Create a function that:
# 1. Takes an image array
# 2. Increases brightness by a given percentage
# 3. Ensures values stay in 0-255 range
# 4. Returns the modified image

def adjust_brightness(image, percentage):
    # Your code here:
    image = image * (percentage / 100)
    image = np.clip(image, 0, 255)  # Keep in valid range
    return image

# Test your function:
test_image = np.random.randint(0, 256, size=(100, 100))
bright_image = adjust_brightness(test_image, 130)  # 30% brighter


print(test_image)

print(f"\n{bright_image}")
```
