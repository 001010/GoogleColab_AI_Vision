```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple pattern to see transformations clearly
pattern = np.array([
    [0, 0, 0, 0],
    [0, 0, 255, 0],
    [255, 255, 255, 255],
    [0, 0, 255, 0]
])

print("Original pattern:")
print(pattern)

# Task 1: Flip vertically (upside down)
flipped_vertical = np.flip(pattern, axis=0)
print("\nFlipped vertically:")
print(flipped_vertical)

flipped_horizontal = np.flip(pattern, axis=1)
print("\nFlipped horizontally:")
print(flipped_horizontal)

rotated_90 = np.rot90(pattern)
print("\nRotated 90 degrees:")
print(rotated_90)

transposed = np.transpose(pattern)
# Or: transposed = pattern.T
print("\nTransposed:")
print(transposed)

fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes[0, 0].imshow(pattern, cmap='gray')
axes[0, 0].set_title('Original')

axes[0, 1].imshow(flipped_vertical, cmap='gray')
axes[0, 1].set_title('Flipped Vertical')

axes[0, 2].imshow(flipped_horizontal, cmap='gray')
axes[0, 2].set_title('Flipped Horizontal')

axes[1, 0].imshow(rotated_90, cmap='gray')
axes[1, 0].set_title('Rotated 90Â°')

axes[1, 1].imshow(transposed, cmap='gray')
axes[1, 1].set_title('Transposed')

axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
```
